```python
import pandas as pd
import numpy as np
from  matplotlib import pyplot as plt
import seaborn as sns
import plotly 
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df=pd.read_csv('E:\/Practice dataset.csv')
```

```python
#No Of Educatated and bechelors#

plt.rcParams['figure.figsize']=(20,10)
plt.bar(df['Education'].value_counts(),df['Education'].value_counts().keys(),width=5.0,color="r")
plt.grid()
plt.show()
```

```python
#No of Income#

plt.rcParams['figure.figsize']=(20,10)
plt.plot(df['Income'].value_counts(),color="r")
plt.grid()
plt.show()
```

```python
#Occupation In Percentage#

df_works=df['Occupation'].value_counts()
df_prof=df['Occupation'].value_counts().keys()

fig=px.pie(data_frame=df,values=df_works,names=df_prof)
fig.show()
```

```python
#Region Of Percentage#

df_region=df['Region'].value_counts()
df_region_count=df['Region'].value_counts().keys()

plt.pie(df_region,labels=df_region_count,autopct="%0.1f%%",shadow=True,startangle=200)
plt.legend()
plt.show()

```

```python
#No of Occupation Earning by Region #

plt.figure(figsize=(15,7))
sns.lineplot(x='Occupation',y='Income',data=df,markers=True,color="indigo",style='Region')
plt.grid()
plt.show()
```

```python
#Gender ration In Percentage #

df_gender=df['Gender'].value_counts()
df_gender_count=df['Gender'].value_counts().keys()

plt.pie(df_gender,labels=df_gender_count,autopct="%0.1f%%",shadow=True,startangle=150,colors=('green','pink'))
plt.legend()
plt.show()
```

```python
#Age and Children#
plt.figure(figsize=(20,15))
sns.barplot(x='Age',y='Children',data=df,)
plt.show()
```

```python
#Male and Female By Education#

plt.rcParams['figure.figsize']=(20,8)
sns.countplot(data=df,x='Gender',hue='Education',palette='ch:.25')
plt.grid()
plt.show()
```
